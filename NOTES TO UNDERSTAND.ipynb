```r
data("Proteomics_Rosettes_CW")
Y <-Metabolomics_Rosettes
X <-Proteomics_Rosettes_CW
#check to see that they come from same samples and everything matches(rows) we don't care if they have same features(cols)
head(data.frame(rownames(X), rownames(Y)))
```

```r
#  4.2 sPLS1 regression y~X
y <- Y[, "Pectin_linearity"]
```

```r
#4.2.1 Number of dimensions using the Q2 criterion
pls1 <- pls(X = X, Y = y, ncomp = 4, mode = 'regression')
set.seed(33)  # For reproducibility with this handbook, remove otherwise
Q2_pls1 <- perf(pls1, validation = 'Mfold', 
                      folds = 10, nrepeat = 5)
plot(Q2_pls1, criterion = 'Q2')
```

```r
#4.2.2 Number of variables to select in X
# Set up a grid of values: 
list.keepX <- c(5:10, seq(15, 50, 5))     

# list.keepX  # Inspect the keepX grid
set.seed(33)  # For reproducibility with this handbook, remove otherwise
tune.spls1.MAE <- tune.spls(X, y, ncomp= 2, 
                            test.keepX = list.keepX, 
                            validation = 'Mfold', 
                            folds = 10,
                            nrepeat = 5, 
                            progressBar = FALSE, 
                            measure = 'MAE')
plot(tune.spls1.MAE)
```

```r
#confirme that we only need one it should be deducted from the garph
choice.ncomp <- tune.spls1.MAE$choice.ncomp$ncomp
# Optimal number of variables to select in X based on the MAE criterion
# We stop at choice.ncomp
choice.keepX <- tune.spls1.MAE$choice.keepX[1:choice.ncomp]  

choice.ncomp
```

```r
choice.keepX
```

```r
#4.2.3 Final sPLS1 model
spls1<- spls(X, y, ncomp = choice.ncomp, keepX = choice.keepX, 
                    mode = "regression")
selectVar(spls1, comp = 1)$X$name   
print(spls1$prop_expl_var$X)     
print(pls1$prop_expl_var$X)
```

```r
#4.2.4 Sample plots
spls1_c2 <- spls(X, y, ncomp = 2, keepX = c(rep(choice.keepX, 2)), 
                   mode = "regression")

plotIndiv(spls1_c2,
          group = Ecotype,
          pch = as.factor(Ecotype),
          legend = TRUE, legend.title = 'Time', legend.title.pch = 'Dose')
#the content of plots is sample(rows) that's why we used ecotype to grp
```
